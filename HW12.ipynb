```python
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets


iris = datasets.load_iris()
X = iris.data
Y = iris.target
target_names = iris.target_names
```

```python
colors = ['navy', 'turquoise', 'darkorange']
```

<b>PCA</b>

```python
from sklearn.decomposition import PCA

X_pca = PCA(n_components=2).fit_transform(X)

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_pca[Y == i, 0], X_pca[Y == i, 1], color=color, label=target_name)
    
plt.legend(bbox_to_anchor=(1, 1), loc=2)
plt.title('PCA of IRIS dataset')
plt.show()
```

<b>t-SNE</b>

```python
from sklearn.manifold import TSNE

for perplexity in [2, 10, 15, 20, 25, 30, 50, 100]:
    X_tsne = TSNE(n_components=2, perplexity=perplexity, n_iter=5000, learning_rate=50).fit_transform(X)

    for color, i, target_name in zip(colors, [0, 1, 2], target_names):
        plt.scatter(X_tsne[Y == i, 0], X_tsne[Y == i, 1], color=color, label=target_name)

    plt.legend(bbox_to_anchor=(1, 1), loc=2)
    plt.title('t-SNE of IRIS dataset with perplexity %d' % (perplexity))
    plt.show()
```
